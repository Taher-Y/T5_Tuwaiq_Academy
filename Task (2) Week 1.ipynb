## 2. Data Visualization

### Create visualizations to illustrate the findings from the EDA.


### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
# Histogram for total crashes
plt.figure(figsize=(10, 6))
sns.histplot(df['Year'], kde=True, bins=20, color='blue')
plt.title('Histogram')

plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Year', y='Kilometers_Driven')
plt.title('Bar Chart')
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(10, 6))
plt.Pie(df, labels= '', autopct='%0.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Scatter Chart
plt.scatter(data=df, x='Year' , y='Kilometers_Driven')
plt.title('Scatter Chart')
plt.show()
```

```python
#Heatmaps Chart
import numpy as np
f= df.drop(['abbrev'], axis=1 )
f1 = f.corr()
```

```python
sns.heatmap(f1, annot=True)
plt.title('Heatmaps Chart')
```
